```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv(r"C:\Users\thoma\Desktop\Code\Formation\Panda\dog_breeds.csv")
df.head()
```

```python
df=df.astype('str')
df.columns=df.columns.astype(str)
df.dtypes
```

```python
df['Height (in)']=df['Height (in)'].apply(lambda x:sum(map(int,x.split('-')))/2)
df['Longevity (yrs)']=df['Longevity (yrs)'].apply(lambda x:sum(map(int, x.split('-')))/2)
```

```python
col=['Height (in)','Longevity (yrs)']
df[col]=df[col].astype(float)
df.dtypes
```
